1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
  return 10*x-5
```

```python
xs = np.linspace(0,10,10)
plt.plot(xs, f(xs), 'o');
```

```python
x = np.linspace(0, 10, 100)
f = 10 * x - 5
f_derivative = np.full_like(x, 10)
plt.figure(figsize=(10, 5))

x_point = 4
f_point = 10 * x_point - 5
df_point = 10
tangent_line = df_point * (x - x_point) + f_point

plt.plot(x, f, label='f(x) = 10x - 5', color="blue")
plt.plot(x, f_derivative, label="f'(x) = 10", color="green")

plt.scatter(x_point, f_point, color='red')
plt.plot(x, tangent_line, '--', color='red', label='Tangent at x = 4')

plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

Висновок: збільшення кількості годин прямопропорційно впливає покращенню оцінки.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
import sympy as sp
```

```python
x = sp.symbols('x')
```

```python

f = sp.sin(x)
df = sp.diff(f, x)
print("Функція:", f)
print("Похідна:", df)
f= sp.lambdify(x, f, 'numpy')
df = sp.lambdify(x, df, 'numpy')

x_sin = np.linspace(-2*np.pi, 2*np.pi, 100)
plt.figure(figsize=(10, 5))

plt.subplot(2, 1, 1)
plt.plot(x_sin, f(x_sin), label='f(x) = sin(x)')
plt.plot(x_sin, df(x_sin), '--', label="f'(x) = cos(x)")
plt.title('Функція sin(x) та її похідна')
plt.legend()
plt.grid(True)
```

```python
f_l = sp.ln(x)
df_l = sp.diff(f_l,x)
print("Функція:", f_l)
print("Похідна:", df_l)

f_l= sp.lambdify(x, f_l, 'numpy')
df_l = sp.lambdify(x, df_l, 'numpy')

x_l = np.linspace(0.1, 5, 100)
plt.figure(figsize=(10, 5))

plt.subplot(2, 1, 1)
plt.plot(x_l, f(x_l), label='f(x) = ln(x)')
plt.plot(x_l, df(x_l), '--', label="f'(x) = 1/x")
plt.title('Функція ln(x) та її похідна')
plt.legend()
plt.grid(True)
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')
f_sym = x**2 - 5*x + 1

df_sym = sp.diff(f_sym, x)

print("f(x) =", f_sym)
print("f'(x) =", df_sym)

f = sp.lambdify(x, f_sym, 'numpy')
df = sp.lambdify(x, df_sym, 'numpy')

x_vals = np.linspace(0, 100, 200)

plt.figure(figsize=(10, 5))

plt.subplot(2, 1, 1)
plt.plot(x_vals, f(x_vals), label='f(x) = x^2 - 5x + 1')
plt.title('Функція загальних витрат')
plt.xlabel('x (кількість одиниць)')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(x_vals, df(x_vals), '--', label="f'(x) = 2x - 5")
plt.title('Похідна (граничні витрати)')
plt.xlabel('x (кількість одиниць)')
plt.ylabel("f'(x)")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
