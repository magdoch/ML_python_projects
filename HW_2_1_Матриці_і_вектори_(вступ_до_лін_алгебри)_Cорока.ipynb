3. Задана матриця `D` розмірністю 2х2. Подумайте спочатку, яку розмірність Ви отрмаєте, коли домножите `result_matrix` на матрицю `D` і який результат буде. А потім виконайте множення і виведіть результат на екран.

```python
A = np.array(
    [[0.18, 0.53],
     [0.75, 0.47],
     [0.35, 0.21]])

B = np.array(
    [[0.17, 0.58, 0.75],
     [0.38, 0.11, 0.15]])

C = np.array([[0], [1]])

D = np.array([[0,1], [1,0]])

display(A, B, C, D)
```

```python
result_matrix = np.dot(B,A)
print(f"Результат множення двох матриць B,A в розмірі {result_matrix.shape}:\n{result_matrix}")
#я так розумію, що нам ж ніхто не заброняє поміняти їх місцями при множені, бо це не вказано в умові тому, щоб отримати заданий розмір я зробила це
```

```python
mult_vector_matrix = np.dot(result_matrix,C)
print(f"Результат множення матриці result_matrix на вектор C:\n{mult_vector_matrix}")
```

```python
mult_matrixs = np.dot(result_matrix,D)
print(f"Результат множення матриці result_matrix на матрицю D:\n{mult_matrixs}")
```
